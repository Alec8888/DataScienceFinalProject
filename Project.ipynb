```python
#import matplotlib
#from matplotlib import pylab
import pandas as pd
```

```python
df = pd.read_csv('./dataset.csv')
df.head()
```

```python
df.describe(include = 'all')
```

```python
print(df.isnull().sum())
```

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create subplots
fig, ax = plt.subplots(2, 1, figsize=(15, 10))

# Plot histogram for popularity
ax[0].hist(df['popularity'], bins=30, edgecolor='black', alpha=0.7)
ax[0].set_title('Histogram of Popularity')
ax[0].set_xlabel('Popularity')
ax[0].set_ylabel('Frequency')

# Plot density plot for popularity using Gaussian Kernel Density Estimate
from scipy.stats import gaussian_kde
density = gaussian_kde(df['popularity'])
x_vals = np.linspace(min(df['popularity']), max(df['popularity']), 1000)
y_vals = density(x_vals)
ax[1].plot(x_vals, y_vals, color='blue')
ax[1].set_title('Density Plot of Popularity')
ax[1].set_xlabel('Popularity')
ax[1].set_ylabel('Density')

# Show the plots
plt.tight_layout()
plt.show()
```

### Popularity Histogram
This shows the frequency of each 'popularity' value. The x-axis represents the 'popularity' and the y-axis represents the frequency of each 'popularity' score. As we can see, a significant number of songs have a popularity score around 0-10, but there's also a peak around the 30-40 range, suggesting that the average song in the dataset has moderate popularity.

### Popularity Density
This plot shows the density of the 'popularity' feature, which can be interpreted as a smoothed version of the histogram. The x-axis represents the 'popularity' and the y-axis represents the density of each 'popularity' score. The peaks of the curve show the values where the songs are most densely populated, which is a similar interpretation to the histogram.

```python
# Create a box plot for popularity
plt.figure(figsize=(10, 5))
plt.boxplot(df['popularity'], vert=False)
plt.title('Box Plot of Popularity')
plt.xlabel('Popularity')
plt.show()
```

### Popularity Box Plot
The box plot shows that the 'popularity' feature has a wide range of values, with several outliers on the higher end of the scale. The median popularity score is around 35, indicating that half of the songs in the dataset have a popularity score below this value and half have a score above this value.

```python
# Compute the correlation matrix for numerical features

numerical_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_df.corr()


# Display the correlation with 'popularity'
correlation_with_popularity = correlation_matrix['popularity'].sort_values(ascending=False)
correlation_with_popularity
```

The correlation coefficients between 'popularity' and the other numerical features in the dataset are relatively small. This means that there's no strong linear relationship between 'popularity' and these features.

- loudness' has the highest positive correlation with 'popularity' (0.050), suggesting that louder songs tend to be more popular, though the relationship is weak.
- 'explicit' has the second highest positive correlation (0.044), implying that songs with explicit content are slightly more likely to be popular.
- 'instrumentalness' has the highest negative correlation with 'popularity' (-0.095), suggesting that songs with higher instrumental content tend to be less popular.

```python
import seaborn as sns

# Filter the DataFrame for explicit and non-explicit songs
explicit_songs = df[df['explicit'] == True]
non_explicit_songs = df[df['explicit'] == False]

# Create subplots
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Plot histogram for explicit songs' popularity
sns.histplot(explicit_songs['popularity'], ax=ax[0], kde=True, color='blue', label='Explicit')
ax[0].set_title('Distribution of Popularity for Explicit Songs')
ax[0].set_xlabel('Popularity')
ax[0].set_ylabel('Density')

# Plot histogram for non-explicit songs' popularity
sns.histplot(non_explicit_songs['popularity'], ax=ax[1], kde=True, color='green', label='Non-Explicit')
ax[1].set_title('Distribution of Popularity for Non-Explicit Songs')
ax[1].set_xlabel('Popularity')
ax[1].set_ylabel('Density')

# Add legend
ax[0].legend()
ax[1].legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import matplotlib.pyplot as plt


average_values = df.groupby('track_genre')[['danceability', 'energy']].mean()
plt.figure(figsize=(10, 6))  
average_values.plot(kind='line')
plt.title('Average Danceability and Energy by Track Genre')
plt.xlabel('Track Genre')
plt.ylabel('Average Value')
plt.tight_layout()
plt.show()
```

### Average Danceability: 

The analysis reveals the average danceability level for different track genres. Danceability is a measure of how suitable a track is for dancing, ranging from 0 to 1. We observed that j-dance has the highest average danceability, indicating that tracks in this genre tend to have a lively and danceable rhythm. On the other hand, ska shows a lower average danceability, suggesting that tracks in this genre may have a more subdued or less danceable rhythm.


### Average Energy: 

The analysis also provides insights into the average energy levels across different track genres. Energy represents the intensity and activity level of a track, also ranging from 0 to 1. It was found that j-dance exhibits the highest average energy, indicating that tracks in this genre are typically energetic and intense. Conversely, dance demonstrates a lower average energy, suggesting that tracks in this genre might be characterized by a more relaxed or mellow vibe.
